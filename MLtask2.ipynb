```python
df.hist(bins=15, figsize=(10, 8), color='skyblue', edgecolor='black')
plt.show()
```

```python
plt.figure(figsize=(22, 5))

for i, column in enumerate(df.select_dtypes(include=['int64', 'float64']).columns, 1):
    plt.subplot(1, len(df.columns), i)
    sns.boxplot(y=df[column], color='lightgreen')
    plt.title(f'Boxplot of {column}')

plt.tight_layout()
plt.show()
```

```python
sns.pairplot(df)
plt.suptitle('Pairplot of Numeric Features', y=1.02)
plt.show()
```

```python
from scipy.stats import zscore

df['Age_zscore'] = zscore(df['Age'])

anomalies = df[(df['Age_zscore'].abs() > 3)]
print(anomalies)
```
